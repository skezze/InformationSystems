4.	Додати до датафрейму поле Age_group як поле Age з градацією у 10 років

```python
mydf=df[['Sex','Age', 'Survived']]
mydf['Age_group']=mydf['Age'].map(lambda x: int(x/10)+1)
#mydf.describe()
mydf.head()
```

5.	Розрахувати кількість пасажирів за кожною групою по полям 'Sex', 'Age_group', 'Survived' за допомогою функції groupby.

```python
surviveddf = mydf[['Sex', 'Age_group', 'Survived']]
#data = surviveddf.groupby(['Sex','Age_group', 'Survived']).size().reset_index()
surviveddf=surviveddf.groupby(['Sex','Age_group', 'Survived']).size().unstack()
surviveddf=surviveddf.fillna(0)
surviveddf.head()
```

6.	Візуалізувати отримані дані за допомогою функції plot.

```python
surviveddf.plot(kind='barh', stacked=True)
```

7.	Визначити кількість врятованих та відсоток відношення врятованих до загиблих для кожноъ групи пасажирів. Визначити групу пасажирыв з максимальною кількістю врятованих та максимальним відсотком відношення врятованих до загиблих.

```python
def calculatePercent(x,y):
    return y/x

max_survival_group = surviveddf
max_survival_group.rename(columns={0:'dies'},inplace=True)
max_survival_group.rename(columns={1:'survived'},inplace=True)
max_survival_group=max_survival_group.fillna(0)
max_survival_group['survivedpercent']= np.divide(max_survival_group['survived'],np.add(max_survival_group['survived'],max_survival_group['dies']))*100
max_survival_group['survived/dies']=np.divide(max_survival_group['survived'],max_survival_group['dies'])
maxsurvived=max_survival_group['survived'].idxmax()
maxsurvivedpercent=max_survival_group['survivedpercent'].idxmax()
print('врятовані пасажири')
print(max_survival_group['survived'])
print('врятовані пасажири в відсотках')
print(max_survival_group['survivedpercent'])
print('группа з максимальною кількістю врятованих пасажирів')
print(maxsurvived)
print('группа з максимальним відсотком врятованих пасажирів')
print(maxsurvivedpercent)

```
